```python
import PIL as pil
from PIL import Image
import numpy as np 
arr = np.zeros((2,2))
arr = np.pad(arr, ((1000,1000), (1000,1000)),'constant',constant_values=(0,0))
for i in range(len(arr)):
    for j in range(len(arr[1])):
        arr[i][j] = (j%2+i%2)%2*255
print(arr)
img = Image.fromarray(np.uint8(arr), "L")
img.save("image.jpg")
```

```python
monaLisa = np.asarray(pil.Image.open('monaLisa.png'))
print(monaLisa)
image = np.zeros((len(monaLisa),len(monaLisa[0])))
for i in range(len(monaLisa)):
    for j in range(len(monaLisa[i])):
        image[i][j] = monaLisa[i][j][0]*0.299+monaLisa[i][j][1]*0.587+monaLisa[i][j][2]*0.114
print(image)
image = Image.fromarray(np.uint8(image),"L")
image.thumbnail((300,300), Image.ANTIALIAS)
image.save("grayLisa.png")
```
