```python
import numpy as np

with open("DailyDelhiClimateTest.csv", "r") as file:
    string = file.read()

array = string.split("\n")
for i in range(len(array)):
    array[i] = array[i].split(",")
array.pop(-1)
newarray = np.array(array)
print(newarray)
```

```python
import numpy as np

with open("DailyDelhiClimateTest.csv", "r") as file:
        array = file.readlines()

for i in range(len(array)):
    array[i] = array[i].split("\n")[0].split(",")
array = np.array(array)
print(array)
```

```python
import numpy as np


with open("DailyDelhiClimateTest.csv", "r") as file:
    array = file.readlines()
    newArray = np.array([array[0]])
    print(newArray)
    for line in array[1:]:
        newArray = np.vstack((newArray, np.array([line.split("\n")[0].split(",")])))
print(newArray)
```
