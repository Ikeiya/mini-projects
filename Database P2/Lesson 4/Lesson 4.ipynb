```python
import PIL as pil
from PIL import Image
import numpy as np 
arr = np.zeros((2000,2000),axis=3)
for i in range(len(arr)):
    for j in range(len(arr[1])):
        arr[i][j] = ((j%2+i%2)%2*255, (j%2+i%2)%2*255, (j%2+i%2)%2*255)
print(arr)
img = Image.fromarray(arr, "RGB")
img.save("image.jpg")
```
