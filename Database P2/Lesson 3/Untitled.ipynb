```python
import numpy as np

with open("DailyDelhiClimateTest.csv", "r") as file:
        array = file.readlines()

for i in range(len(array)):
    array[i] = array[i].split("\n")[0].split(",")
array = np.array(array)

title = array[0]
array = np.delete(array, 0, 0)

print(array)
```

```python
date = []
temp = []
humidity = []
windSpeed = []
pressure = []

date.append([datetime.strptime(i[0], '%Y-%m-%d').date() for i in array])
temp.append([float(i[1]) for i in array])
humidity.append([float(i[2]) for i in array])
windSpeed.append([float(i[3]) for i in array])
pressure.append([float(i[4]) for i in array])

print(pressure)
```

```python
fig, axs = plt.subplots(2,2,figsize=(20, 3))
fig.tight_layout()
axs[0][0].set_title("Temperature")
axs[0][0].plot(date[0], temp[0])
axs[0][1].set_title("Humidity")
axs[0][1].plot(date[0], humidity[0])
axs[1][0].set_title("Wind Speed")
axs[1][0].plot(date[0], windSpeed[0])
axs[1][1].set_title("Pressure")
axs[1][1].plot(date[0], pressure[0])
plt.show()
```
